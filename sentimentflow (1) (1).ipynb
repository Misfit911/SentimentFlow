#### 4.1.1 Corrupted Records in the *tweet_text* Column

The data seems to be corrupted in some records. To correct this issue, we create a function that can find these corrupted records, and returns their indexes. We use this index to remove the record from the working dataframe.

```python
# A function targeting corrupted records
def is_corrupted(tweet):
    """This func returns the index of any record that is corrupted"""
    corrupted_cols = []
    for key, text in enumerate(tweet):
        if any(ord(char) > 127 for char in str(text)) == True:
            corrupted_cols.append(key)
    return corrupted_cols
```

## 6.0 TEXT PREPROCESSING

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download required NLTK data
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Initialize stopwords and lemmatizer
stop_words = set(stopwords.words('english') + ['sxsw', 'sxswi', 'link', 'rt'])
lemmatizer = WordNetLemmatizer()

def lemmatize_tweet(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove mentions
    text = re.sub(r'@\w+', '', text)
    # Remove hashtags (keep the text after the #)
    text = re.sub(r'#', '', text)
    # Remove special characters like "&quot;"
    text = re.sub(r'&\w+;', '', text)
    # Remove punctuation
    text = "".join([char for char in text if char not in string.punctuation])
    # Tokenize text
    tokens = re.split('\W+', text.lower())
    # Remove stopwords and lemmatize the tokens
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    # Remove extra white spaces
    tokens = [word.strip() for word in tokens if word.strip() != '']
    # Remove numbers
    tokens = [word for word in tokens if not word.isdigit()]
    # Tag parts of speech
    pos_tags = nltk.pos_tag(tokens)
    # Filter tokens to retain only nouns, adjectives, verbs, and adverbs
    important_pos = {'NN', 'NNS', 'NNP', 'NNPS', 'JJ', 'JJR', 'JJS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'RB', 'RBR', 'RBS'}
    tokens = [word for word, tag in pos_tags if tag in important_pos]
    return tokens

# Apply the clean_text function to the 'tweet' column
df['lemmatized_tweet'] = df['tweet'].apply(lambda x: lemmatize_tweet(x))

def join_text(tokens):
    """
    This function takes a list of tokens (words) and returns them as a single string.
    Each token is separated by a space.

    Parameters:
    tokens (list of str): A list of tokens to be joined.

    Returns:
    str: The tokens joined into a single string separated by spaces.
    """
    return " ".join(tokens)

df['clean_tweet'] = df['lemmatized_tweet'].apply(lambda x: join_text(x))

# Print the cleaned tweets
df.head()
```

### ***Creating Bigrams***

Analyzing bigrams improves context understanding by considering pairs of words together, which helps in understanding phrases that might have specific meanings different from their individual words.

```python
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures


# Alias for BigramAssocMeasures
bigram_measures = BigramAssocMeasures()

# Create BigramCollocationFinder
finder = BigramCollocationFinder.from_words(flattened_lemmatized_tweet)

# Score bigrams by raw frequency
scored = finder.score_ngrams(bigram_measures.raw_freq)

# Display the 20 most common bigrams
for bigram, score in scored[:20]:
    print(bigram, score)
```

### TRAIN TEST SPLIT

Train_test_split split the data into train and test sets.X_train and y_train are the features and target variable for the training set.X_test and y_test are the features and target variable for the test set.
We will use the test_size of 0.3 which specifies that 30% of the data should be used for testing and a random_state of 42 used to ensure that the split is reproducible;

```python
# The target variable `emotion` column
y = df['emotion']
# The features column
X = df['clean_tweet']
```

```python
# Splitting the dataset into train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### VECTORIZATION

Vectorization in Natural Language Processing (NLP) is the process of converting text data into numerical vectors that can be used by machine learning algorithms. Since machine learning models require numerical input, text data must be transformed into a format they can understand.There are several techniques of Vectorization but in this case We will tfidf vectorizer.TF-IDF (Term Frequency-Inverse Document Frequency) is a popular vectorization method in NLP used to transform text data into numerical vectors, which helps in identifying the importance of a word in a document relative to a collection of documents (corpus). It combines two statistics: Term Frequency (TF) and Inverse Document Frequency (IDF).

```python
# Set the max_features to 50000
# Instantiate the tfidf vectorizer
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1,2), max_features=50000)
tfidf_vectorizer.fit(X_train)
X_train_tfidf = tfidf_vectorizer.transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
print(X_train_tfidf.shape)
print(X_test_tfidf.shape)
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_tfidf, y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Define the models
Bayes = MultinomialNB()
LogReg = LogisticRegression(max_iter=10000)
RForest = RandomForestClassifier()
svc = SVC()
SVM_linear = LinearSVC(max_iter=10000)


Models = [Bayes, LogReg, RForest, svc, SVM_linear]
Models_Dict = {0: "Naive Bayes", 1: "Logistic Regression", 2: "Random Forest", 3: "SVC", 4: "SVM_linear"}

# Cross-validate models
print("Cross-validation Results:")
for i, model in enumerate(Models):
    accuracy = cross_val_score(model, X_test_tfidf, y_test, cv=5, scoring="accuracy").mean()
    print("{} Test Accuracy: {:.4f}".format(Models_Dict[i], accuracy))
```

```python
from sklearn.metrics import accuracy_score

# Train and evaluate models
print("\nFinal Model Evaluation on Test Data:")
for i, model in enumerate(Models):
    model.fit(X_train_resampled, y_train_resampled)  # Train the model
    y_pred = model.predict(X_test_tfidf)  # Predict on the test data
    accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy
    print("{} Test Accuracy: {:.4f}".format(Models_Dict[i], accuracy))
```

#### MODEL EVALUATION :

**Cross-validation Results:** SVM_linear outperforms other models, followed by Logistic Regression, SVC, Random Forest, and Naive Bayes.

**Final Test Accuracy:** SVC achieved the highest accuracy, with Random Forest and SVM_linear following. Logistic Regression and Naive Bayes are less accurate.
